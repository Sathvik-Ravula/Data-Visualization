```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## *immigrants_emigrants_by_age.csv*

# **Loading Dataset**

```python
df = pd.read_csv('immigrants_emigrants_by_age.csv')
```

```python
df.head()
```

```python
dist_specific_Nationality = df[(df["District Name"] == "Ciutat Vella")]
dist_specific_Nationality
```

```python
AgeCount = {}
for i in dist_specific_Nationality['Age'].unique() :
    if(i == '15-19' or i == '20-24' or i == '25-29' or i == '30-34'):
        AgeCount[i] = dist_specific_Nationality[dist_specific_Nationality['Age']==i].Immigrants
print(AgeCount)
```

```python
plt.figure(figsize=(10,10))
plt.title('No Of Immigrants Between 15-34 Years In Ciutat Vella',fontsize=24)
plt.xlabel('Age Group',fontsize=17)
plt.tick_params(labelsize=15)
plt.ylabel('Frequency',fontsize=17)
plt.boxplot(x = [x for x in AgeCount.values()], labels=[x for x in AgeCount.keys()])
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.rcParams["figure.figsize"] = [7.50, 3.50]

plt.hist(x = AgeCount.values(), label = list(AgeCount.keys()))
plt.legend(fontsize=15)
plt.title('No Of Emigrants Between 15-34 Years In Ciutat Vella',fontsize=24)
plt.xlabel('Immigrants',fontsize=17)
plt.tick_params(labelsize=15)
plt.ylabel('Frequency',fontsize=17)
plt.show()
```
